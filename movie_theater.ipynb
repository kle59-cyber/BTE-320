```python
# Movie theater profit table

ticket_price = 5
advertising_cost = 30

for tickets in range(1, 26):   # 1 to 25 tickets
    revenue = tickets * ticket_price
    profit = revenue - advertising_cost
    print(tickets, profit)
```
