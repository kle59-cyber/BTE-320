```python
# Simple income statement: sales, COGS, gross profit

sales = float(input("Enter sales ($): "))

# COGS is 40% of sales
cogs = sales * 0.4

gross_profit = sales - cogs

print("Sales: $", sales)
print("Cost of Goods Sold: $", cogs)
print("Gross Profit: $", gross_profit)

```
