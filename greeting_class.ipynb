```python
# Greeting class

class Greeting:
    def __init__(self, name):
        self.name = name

    def say_hello(self):
        print("Hello,", self.name + "!")


def main():
    user_name = input("Enter your name: ")
    g = Greeting(user_name)
    g.say_hello()


if __name__ == "__main__":
    main()
```
