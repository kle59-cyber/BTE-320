```python
# Append items to a list iteratively (while loop)

numbers = []   # start with an empty list

value = float(input("Enter a number (0 to stop): "))

while value != 0:
    numbers.append(value)
    value = float(input("Enter a number (0 to stop): "))

# sort in descending order
numbers.sort(reverse=True)

print("Numbers in descending order:", numbers)
```
