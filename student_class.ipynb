```python
# Student class

class Student:
    def __init__(self, name, major, gpa):
        self.name = name
        self.major = major
        self.gpa = gpa

    def get_info(self):
        return self.name + " is majoring in " + self.major + " with a GPA of " + str(self.gpa)

def main():
    name = input("Enter student name: ")
    major = input("Enter major: ")
    gpa = float(input("Enter GPA: "))

    s = Student(name, major, gpa)
    print(s.get_info())

if __name__ == "__main__":
    main()
```
