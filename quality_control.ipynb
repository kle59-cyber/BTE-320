```python
# Quality control manager: average weight with exception handling

weights = []
done = False

while not done:
    try:
        value = float(input("Enter weight (0 to stop): "))

        if value == 0:
            done = True
        else:
            weights.append(value)

    except ValueError:
        print("Invalid input. Please enter a number.")

# After loop ends, compute average
if len(weights) > 0:
    avg = sum(weights) / len(weights)
    print("Average weight:", avg)
else:
    print("No weights entered.")
```
