```python
# Product Prices (Dictionaries)

# dictionary of product IDs and prices
products = {
    "P-01": 9.95,
    "P-02": 10.50
}

prod_id = input("Enter product ID: ")

if prod_id in products:
    print(prod_id, products[prod_id])
else:
    print(prod_id, "not found")
```
